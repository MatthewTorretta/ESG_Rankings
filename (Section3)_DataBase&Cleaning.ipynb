## Cleaning

```python
selected_columns = 

df[[ "Fund profile: Shareclass name", "Fund profile: Ticker", \
     "Fund profile: Shareclass type", 
     "Fund profile: Fund net assets", \
     "Fossil Free Funds: Fossil fuel grade", \
     "Fund profile: Sustainability mandate", \ 
     "Fossil Free Funds: Fossil fuel holdings, weight", \
     "Fossil Free Funds: Relative carbon footprint (tonnes CO2 / $1M USD invested)", \
     "Fossil Free Funds: Clean200, weight"]]
```

```python
clean_df = selected_columns.copy()
clean_df.dropna()
clean_df["Fund profile: Sustainability mandate"] = clean_df["Fund profile: Sustainability mandate"].fillna('N')
```

```python
color_mapping = {'A': 'green', 'B': 'lightgreen', 'C': 'yellow', 'D': 'orange', 'F': 'red'}
FFF = clean_df[0:51].style.\
                    applymap(lambda v: f"background-color: {color_mapping.get(v, '')}") \
                    .format({
                    'Fossil Free Funds: Fossil fuel holdings, weight': '{:,.2%}'.format,
                    'Fossil Free Funds: Clean200, weight': '{:,.2%}'.format,
                    'Fund profile: Fund net assets': '${:,.2f}'.format,
                    'Fossil Free Funds: Relative carbon footprint (tonnes CO2 / $1M USD invested)': '{:,.2f}'.format,
                            })
FFF
```

 ## DATABASE Creation

```python
con = lite.connect('FossilFreeFunds.db')
clean_df.to_sql(name="FFF", con = con, if_exists = 'replace', index = False)
con.commit()
cur = con.cursor()
```

```python
Qry = '''SELECT * FROM FFF'''
FFF = pd.read_sql_query(Qry, con)
FFF.head(10)
```

## Top ten Sustainable funds

```python
 Qry = '''SELECT * FROM FFF 
        WHERE "Fossil Free Funds: Fossil fuel grade" =="A" AND
        "Fund profile: Sustainability mandate" == "Y"
        ORDER BY 
        "Fossil Free Funds: Fossil fuel holdings, weight" ASC,
        "Fund profile: Fund net assets" DESC,
        "Fossil Free Funds: Relative carbon footprint (tonnes CO2 / $1M USD invested)" ASC,
        "Fossil Free Funds: Clean200, weight" DESC; '''

Top10 = pd.read_sql_query(Qry, con)
Top10.head(10)

#cur.execute(Qry)
#result = cur.fetchall()
#result
```

## Bottom Ten funds 

```python
Qry = '''SELECT * FROM FFF 
        WHERE "Fund profile: Ticker" not like 'None' AND
        "Fossil Free Funds: Fossil fuel grade" =="F" AND
        "Fund profile: Sustainability mandate" == "N"
        ORDER BY 
        "Fossil Free Funds: Fossil fuel holdings, weight" DESC,
        "Fund profile: Fund net assets" DESC,
        "Fossil Free Funds: Relative carbon footprint (tonnes CO2 / $1M USD invested)" DESC,
        "Fossil Free Funds: Clean200, weight" DESC; '''

Bottom10 = pd.read_sql_query(Qry, con)
Bottom10.head(10)
```

## Base Middle Fund

```python
Qry = '''SELECT AVG("Fossil Free Funds: Fossil fuel holdings, weight")
            FROM (SELECT "Fossil Free Funds: Fossil fuel holdings, weight"
                  FROM FFF
                  ORDER BY "Fossil Free Funds: Fossil fuel holdings, weight"
                  LIMIT 2 - (SELECT COUNT(*) FROM FFF) % 2    -- odd 1, even 2
                  OFFSET (SELECT (COUNT(*) - 1) / 2
                          FROM FFF))'''

cur.execute(Qry)
result = cur.fetchall()
result
```

```python
Qry = '''SELECT * FROM FFF ORDER BY ABS(0.054376 - "Fossil Free Funds: Fossil fuel holdings, weight") LIMIT 1 '''
Median = pd.read_sql_query(Qry, con)
Median
```
